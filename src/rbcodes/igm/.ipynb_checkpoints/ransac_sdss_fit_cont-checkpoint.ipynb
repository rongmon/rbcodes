```python
def data_loader(rel_path='/', filename='', dtype=float):
    '''Load any data previously stored as numpy array

    Parameters:
        str rel_path = relative path to the working folder
        str filename = file name
        dtype = data type
    Return:
        numpy array
    '''
    print(filename+'  is loading...')
    return np.loadtxt(rel_path+filename, dtype=dtype, delimiter=',')

def data_saver(varname, rel_path='/', filename=''):
    '''Load any data previously stored as numpy array

    Parameters:
        str rel_path = relative path to the working folder
        str filename = file name
        dtype = data type
    Return:
        numpy array
    '''
    print(filename+'  is saving...')

    if type(varname[0]) == np.str_:
        return np.savetxt('./'+rel_path+filename, varname, delimiter=',', fmt='%s')
    else:
        return np.savetxt('./'+rel_path+filename, varname, delimiter=',')

def split_specs(wave, spec, split_idx=1216):
    #red_idx = np.where(wave> split_idx)[0]
    #blue_idx = np.where(wave <= split_idx)[0]
    red_spec = spec[:, wave>split_idx]
    blue_spec = spec[:, wave <= split_idx]
    return blue_spec, red_spec

def chop_specs(wave, spec, start_idx=0):
    if start_idx==0:
        print('Nothing is chopped off.')
        return spec
    else:
        return spec[:,wave>start_idx], wave[wave>start_idx]
```

#Fitting continua

```python
working_folder = './Project_2_reionization/SDSS_DR16_z=2_data/'
sdss_good = data_loader(working_folder, 'sdss_spec_good.csv')
sdss_name_good = data_loader(working_folder, 'sdss_name_good.csv', dtype=str)
sdss_z_good = data_loader(working_folder, 'sdss_z_good.csv')
sdss_sn_good = data_loader(working_folder, 'sdss_sn_good.csv')
sdss_err_good = data_loader(working_folder, 'sdss_err_good.csv')
sdss_wave = data_loader(working_folder, 'sdss_wave.csv')
sdss_ivar_good = data_loader(working_folder, 'sdss_ivar_good.csv')
```

####1. median filter of size X
X is user-defined parameter, must be odd number

```python
sdss_md = [medfilt(sdss_good[i], 149) for i in range(len(sdss_good))]
```

```python
# randomly select 10 to display
# keep the random indices throughout to see the final result
rmd_idx = np.random.randint(0, len(sdss_md), 10)
for i in rmd_idx:
    plt.figure(figsize=(10,6))
    plt.plot(sdss_wave, sdss_good[i], c='grey', label='SDSS Spectrum')
    plt.plot(sdss_wave, sdss_md[i], c='blue', label='RunningMedFilter, 49')
    plt.xlabel('Wavelength ($\AA$)', fontsize=15)
    plt.ylabel('Relative Flux (at ~1280$\AA$)', fontsize=15)
    plt.title('SDSS Spectrum at z={:.3f}'.format(sdss_z_good[i]), fontsize=20)
    plt.legend(loc='upper right', fontsize=12)
    plt.ylim([-0.01, sdss_good[i].max()+2])
    plt.show()
```

####2. Local maxima

```python
# for 1 spectrum:
# (peaks, prop) = find_peaks(sdss_good, height=sdss_md)
loc_max = [find_peaks(sdss_good[i], height=sdss_md[i]) for i in range(len(sdss_good))]
```

```python
sdss_good.shape, len(loc_max), sdss_wave.shape
```

```python
# loc_max[0]
#       -- array: positions of local maxima
#       -- 'peak_heights: local flux maxima

# randomly select 10 to display
for i in rmd_idx:
    plt.figure(figsize=(10,6))
    plt.plot(sdss_wave, sdss_good[i], c='grey', label='SDSS Spectrum')
    plt.plot(sdss_wave, sdss_md[i], c='blue', label='RunningMedFilter, 49')
    plt.scatter(sdss_wave[loc_max[i][0]], sdss_good[i][loc_max[i][0]], color='red', label='Local Maxima')
    plt.xlabel('Wavelength ($\AA$)', fontsize=15)
    plt.ylabel('Relative Flux (at ~1280$\AA$)', fontsize=15)
    plt.title('SDSS Spectrum at z={:.3f}'.format(sdss_z_good[i]), fontsize=20)
    plt.legend(loc='upper right', fontsize=12)
    plt.ylim([-0.01, sdss_good[i].max()+1])
    plt.show()
```

####3. form an upper evenlope

```python
peak_interp = [np.interp(sdss_wave, 
                         sdss_wave[loc_max[i][0]],
                         sdss_good[i][loc_max[i][0]]) \
               for i in range(len(sdss_good))]
peak_interp = np.array(peak_interp)
peak_interp.shape
```

```python
for i in rmd_idx:
    plt.figure(figsize=(10,6))
    plt.plot(sdss_wave, sdss_good[i], c='grey', label='SDSS Spectrum')
    plt.plot(sdss_wave, sdss_md[i], c='blue', label='RunningMedFilter, 49')
    plt.scatter(sdss_wave, peak_interp[i], color='orange', label='Upper Evenlope')
    plt.xlabel('Wavelength ($\AA$)', fontsize=15)
    plt.ylabel('Relative Flux (at ~1280$\AA$)', fontsize=15)
    plt.title('SDSS Spectrum at z={:.3f}'.format(sdss_z_good[i]), fontsize=20)
    plt.legend(loc='upper right', fontsize=12)
    plt.ylim([-0.01, sdss_good[i].max()+1])
    plt.show()
```

####4. RANSAC
RANSAC(spec-evenlope)

```python
sp_diff = sdss_good - peak_interp

for i in rmd_idx:
    plt.figure(figsize=(10,6))
    plt.plot(sdss_wave, sp_diff[i], c='black', label='Flux Difference')
    plt.xlabel('Wavelength ($\AA$)', fontsize=15)
    plt.ylabel('Flux Diff[Spec-UpperEvenlope] (at ~1280$\AA$)', fontsize=15)
    plt.title('SDSS Spectrum at z={:.3f}'.format(sdss_z_good[i]), fontsize=20)
    plt.legend(loc='upper right', fontsize=12)
    plt.ylim([sp_diff[i].min()-.1, sp_diff[i].max()+.1])
    plt.show()
```

```python
inlier_masks = []
outlier_masks = []
for i in range(len(sp_diff)):
    ransac = RANSACRegressor()
    ransac.fit(sdss_wave.reshape(-1,1), sp_diff[i])
    inlier_masks.append(ransac.inlier_mask_)
    outlier_masks.append(np.logical_not(ransac.inlier_mask_))
```

```python
for i in rmd_idx:
    plt.figure(figsize=(10,6))
    plt.plot(sdss_wave, sp_diff[i], c='black', label='Flux Difference')
    plt.plot(sdss_wave[inlier_masks[i]], sp_diff[i][inlier_masks[i]], c='green', label='Inliers')
    plt.xlabel('Wavelength ($\AA$)', fontsize=15)
    plt.ylabel('Flux Diff[Spec-UpperEvenlope] (at ~1280$\AA$)', fontsize=15)
    plt.title('SDSS Spectrum at z={:.3f}'.format(sdss_z_good[i]), fontsize=20)
    plt.legend(loc='upper right', fontsize=12)
    plt.ylim([sp_diff[i].min()-.1, sp_diff[i].max()+.1])
    plt.show()
```

####5. Use `inlier_masks` to interpolate

```python
spec_inliers = [np.interp(sdss_wave, 
                         sdss_wave[inlier_masks[i]],
                         sdss_good[i][inlier_masks[i]]) \
               for i in range(len(sdss_good))]
spec_fit = [medfilt(spec_inliers[i], 99) for i in range(len(spec_inliers))]
```

```python
for i in rmd_idx:
    plt.figure(figsize=(10,6))
    plt.plot(sdss_wave, sdss_good[i], c='grey', label='SDSS Spectrum')
    plt.plot(sdss_wave, sdss_md[i], c='blue', label='RunningMedFilter, 49')
    plt.plot(sdss_wave, spec_fit[i], c='red', label='RANSAC fit')
    plt.xlabel('Wavelength ($\AA$)', fontsize=15)
    plt.ylabel('Relative Flux (at ~1280$\AA$)', fontsize=15)
    plt.title('SDSS Spectrum at z={:.3f}'.format(sdss_z_good[i]), fontsize=20)
    plt.legend(loc='upper right', fontsize=12)
    plt.ylim([-0.01, sdss_good[i].max()+2])
    plt.show()
```

####6. Save the result

```python
data_saver(spec_fit, working_folder, 'sdss_fit.csv')
```
